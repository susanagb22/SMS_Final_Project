**Smart Trash Containers Data Set**

A program startedin 2017 has lead to a large increase in smart trash containers in the neighborhoods of pittsburgh as of 2019. These trash containers are public waste bins found around the neighborhoods that have fill-level sensors which communicate when the bin is nearing capacity and sends an alert to the city's Department of Public Works (DPW). Because trash containers are now being maintained better, there is a large reduce in litter and pests since there isn't trash overfilling the containers. This therefore provides cleaner public spaces and improved hygiene for the neighborhood. These alerts also help with lowering the emissions of greenhouse gases since the DPW is able to know when to change a trash container rather than following a fixed schedule every day.

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import csv
```

```python
df = pd.read_csv('SmartTrashContainers.csv')

trash_counts = df['neighborhood'].value_counts()

print(f"Total neighborhoods: {trash_counts.size}")

for neighborhood, count in trash_counts.items():
    print(f"{neighborhood}: {count}")

top_10 = trash_counts.head(10)
```

```python
plt.bar(top_10.index, top_10.values, color='green') #change color to green
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Smart Trash Containers')
plt.title('Top 10 Neighborhoods With the Most Smart Trash Containers in Pittsburgh')
plt.xticks(rotation=45)
plt.show()
```

The bar graph above reprsents the top 10 neighborhoods in Pittsburgh that have the most smart trash containers with Shadyside having a total of 76 trash containers. These containers are very good for keeping the public and enviornment safe so the more there are, the better!
